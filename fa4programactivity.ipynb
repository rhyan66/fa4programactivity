```python
num = 1
while num <= 20:
    if num % 2 == 0:
        num += 1
        continue
    print(num)
    num += 1
```

```python
correct_password = "Chicken"

while True:
    guess = input("Enter the password: ")

    if guess == correct_password:
        print("Access Granted!")
        break

    print("Incorrect password. Try again.")
```

```python
for num in range(1, 11):

    if num == 8:
        break

    if num == 4:
        continue

    if num == 6:
        pass

    print(num)
```
